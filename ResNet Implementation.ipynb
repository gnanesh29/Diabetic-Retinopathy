```python
train_datagen=ImageDataGenerator(
              rescale=1./255,
              shear_range=0.2,
              rotation_range=20,
              validation_split=0.2
)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
train_generator = train_datagen.flow_from_directory(
    directory='./data',
    target_size=(28, 28),
    color_mode="rgb",
    class_mode="categorical",
    batch_size=32,
    subset='training')
```

```python
validation_generator = train_datagen.flow_from_directory(
    directory='./data',
    target_size=(28, 28),
    color_mode="rgb",
    class_mode="categorical",
    batch_size=32,
    subset='validation')
```

```python
test_generator = test_datagen.flow_from_directory(
    directory='./data',
    target_size=(28, 28),
    color_mode="rgb",
    class_mode="categorical",
    batch_size=32)
```

```python
def res_block(X, filters):
    X_copy = X
    f1, f2, f3 = filters
    X = Conv2D(f1, (1, 1), strides=(1, 1))(X)
    X = MaxPooling2D((2,2))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = Conv2D(f2, kernel_size=(3, 3), strides=(1, 1), padding='same')(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = Conv2D(f3, kernel_size=(1, 1), strides=(1, 1))(X)
    X = BatchNormalization(axis=3)(X)
    X_copy = Conv2D(f3, kernel_size=(1, 1), strides=(1, 1))(X_copy)
    X_copy = MaxPooling2D((1, 1))(X_copy)
    X_copy = BatchNormalization(axis=3)(X_copy)
    X = Add()([X, X_copy])
    X = Activation('relu')(X)
    X_copy = X
    X = Conv2D(f1, (1, 1), strides=(1, 1))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = Conv2D(f2, kernel_size=(3, 3), strides=(1, 1), padding='same')(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = Conv2D(f3, kernel_size=(1, 1), strides=(1, 1))(X)
    X = BatchNormalization(axis=3)(X)
    X = Add()([X, X_copy])
    X = Activation('relu')(X)
    X_copy = X
    X = Conv2D(f1, (1, 1), strides=(1, 1))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = Conv2D(f2, kernel_size=(3, 3), strides=(1, 1), padding='same')(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    X = Conv2D(f3, kernel_size=(1, 1), strides=(1, 1), padding='same')(X)
    X = BatchNormalization(axis=3)(X)
    X = Add()([X, X_copy])
    X = Activation('relu')(X)
    return X


```

```python
input_shape = (28,28,3)
X_input = Input(input_shape)
X = ZeroPadding2D((3,3))(X_input)
X = Conv2D(3, (3,3), strides= (2,2))(X)
X = BatchNormalization(axis =3)(X)
X = Activation('relu')(X)
X = AveragePooling2D((2,2))(X)
X = Flatten()(X)
X = Dense(2, activation = 'softmax')(X)
model = Model( inputs= X_input, outputs = X, name = 'Resnet18')
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics= ['accuracy'])
```

```python
earlystopping = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=3)
```

```python
checkpointer = ModelCheckpoint(filepath="dr_28_5.hdf5", verbose=1, save_best_only=True)
```

```python
hist = model.fit(train_generator, steps_per_epoch = train_generator.n // 32, epochs = 5, validation_data= validation_generator, validation_steps= validation_generator.n // 32, callbacks=[checkpointer , earlystopping])
```

```python
hist.history
```

```python
print(hist.history.keys())
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

```python
evaluate = model.evaluate(test_generator, steps = test_generator.n // 32, verbose =1)

print('Accuracy Test : {}'.format(evaluate[1]))
```
