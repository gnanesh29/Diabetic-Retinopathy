```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# Load the image
img = Image.open('224.jpeg')  # Replace 'path_to_your_image.jpg' with the path to your image
img_gray = img.convert('L')  # Convert to grayscale

# Convert the image to a numpy array
img_array = np.array(img_gray)

# Perform 2D FFT
fft_img = np.fft.fft2(img_array)
fft_shifted = np.fft.fftshift(fft_img)

# Calculate amplitude and phase
amplitude = np.abs(fft_shifted)
phase = np.angle(fft_shifted)

# Reconstruct the image using amplitude and phase
reconstructed_img = np.fft.ifft2(np.fft.ifftshift(amplitude * np.exp(1j * phase))).real

# Display original image
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.imshow(img_array, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display amplitude
plt.subplot(132)
plt.imshow(np.log(amplitude + 1), cmap='gray')  # Adding 1 to avoid log(0)
plt.title('Amplitude')
plt.axis('off')

# Display phase
plt.subplot(133)
plt.imshow(phase, cmap='gray')
plt.title('Phase')
plt.axis('off')

plt.show()



# Display reconstructed image
plt.figure(figsize=(6, 6))
plt.imshow(reconstructed_img, cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

def sinusoidal_waveform(image_path, frequency, amplitude, phase):
    # Open the image
    img = Image.open(image_path)

    # Convert the image to grayscale
    img_gray = img.convert('L')

    # Convert the image to a numpy array
    img_array = np.array(img_gray)

    # Get the dimensions of the image
    height, width = img_array.shape

    # Create an array of x values (corresponding to the width of the image)
    x = np.arange(width)

    # Create a sinusoidal waveform with phase shift
    sinusoidal_wave = amplitude * np.sin(2 * np.pi * frequency * x / width + phase)

    # Add the sinusoidal waveform to each row of the image
    for i in range(height):
        img_array[i] = np.clip(img_array[i] + sinusoidal_wave.astype(np.uint8), 0, 255)

    # Create a new image from the modified pixel array
    img_wave = Image.fromarray(img_array)

    # Plot the original image
    plt.figure(figsize=(10, 5))
    plt.subplot(2, 1, 1)
    plt.imshow(img_gray, cmap='gray')
    plt.title('Original Image')
    plt.axis('off')

    # Plot the sinusoidal waveform
    plt.subplot(2, 1, 2)
    plt.plot(x, sinusoidal_wave, color='red')
    plt.title('Sinusoidal Waveform')
    plt.xlabel('Pixel Position')
    plt.ylabel('Amplitude')
    plt.grid(True)

    plt.show()

# Example usage
image_path = '224.jpeg'  # Path to your image file
frequency = 3  # Frequency of the sinusoidal waveform
amplitude = 4611671.0  # Amplitude of the sinusoidal waveform
phase = 0  # Phase shift in radians

sinusoidal_waveform(image_path, frequency, amplitude, phase)
```
